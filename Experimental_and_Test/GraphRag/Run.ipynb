## Now we have the project initialized and loaded from dataset we have. lets test:

```python
import pandas as pd

# Define the file path
csv_file_path = '/home/zihan/Desktop/Manufacturing_QA/Test/ch24V2.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)
df
```

```python
import subprocess
import pandas as pd


# Initialize an empty list to store outputs
questions = df['Question'].tolist()
results = []

for question in questions:
    # Convert the question to a string (if not already)
    question_str = str(question)
    
    # Run the script and capture the output
    result = subprocess.run(['python', '-m', 'graphrag.query', '--root', '.', '--method', 'global', question_str], capture_output=True, text=True)
    
    # Access the captured output
    output = result.stdout
    error_output = result.stderr
    
    # Print the outpu for debugging
    print(output)
    if error_output:
        print("Errors:", error_output)
    
    # Append the output to the list
    results.append(output)

# Now 'outputs' contains the results for all questions
print(results)
```

```python
results
```

```python
copyoutputs = results.copy()
```

```python
import csv

with open('TestResult.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['log_entry'])  # Write the header
    for entry in copyoutputs:
        writer.writerow([entry.strip()])  # Write each log entry as a row

print("Log entries saved to copyoutputs.csv")
```

```python
import re
# Function to extract "Global Search Response" content
def extract_global_search_response(log_entry):
    match = re.search(r'SUCCESS: Global Search Response: (.*)', log_entry, re.DOTALL)
    if match:
        return match.group(1).strip()
    return None

# Extract content from each log entry
extracted_contents = [extract_global_search_response(entry) for entry in copyoutputs if extract_global_search_response(entry) is not None]
extracted_contents
```

```python
len(extracted_contents)
```

```python
Outputs_df = pd.DataFrame(extracted_contents, columns=['GraphRag'])
df['GraphRag'] = Outputs_df['GraphRag']

df
```
