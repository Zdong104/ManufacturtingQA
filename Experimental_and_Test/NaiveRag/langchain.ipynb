```python
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator="\n\n",
    chunk_size=300,
    chunk_overlap=100,
    length_function=len,
    is_separator_regex=False,
)
```

```python
chunks = text_splitter.split_documents(documents)
```

```python
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma


# Create the Chroma index from documents and save it to a SQLite database file
db_file_path = '/home/zihan/Desktop/Manufacturing_QA/Experimental_and_Test/NaiveRag/index.db'
index = Chroma.from_documents(chunks, OpenAIEmbeddings(openai_api_key = api_key), persist_directory=db_file_path)


## Run Below to not save the files.
# index = Chroma.from_documents(chunks, OpenAIEmbeddings(openai_api_key = api_key))
```

```python
retriever = index.as_retriever()
```

```python
from operator import itemgetter

template = """
Answer the question based only on the following context:
{context}

Answer the following question:
Question: {question}
"""
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

prompt = ChatPromptTemplate.from_template(template)
model = ChatOpenAI(model_name="gpt-4o", api_key=api_key)
```

```python
rag_chain = (
    {
        "context": itemgetter("question") | retriever,
        "question": itemgetter("question")
    }
    | prompt
    | model
    | StrOutputParser()
)
```

```python
result = rag_chain.invoke({"question": "Define manufacturing."})
```

```python
print(result)
```

```python
import pandas as pd

# Define the file path
csv_file_path = '/home/zihan/Desktop/Manufacturing_QA/Experimental_and_Test/GraphRag/TestResult.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)
questions = df['Question'].tolist()
results = []
for question in questions:
    result = rag_chain.invoke({"question": question})
    print(result)
    results.append(result)

Outputs_df = pd.DataFrame(results, columns=['LangChain'])
Outputs_df
    
```
